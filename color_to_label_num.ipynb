```python
import os
import numpy as np
from PIL import Image
from collections import defaultdict
import shutil
```

```python
mask_dir = "/content/drive/MyDrive/classification_dataset/projet/dataset_COCO/train/augm_mask"   # Dossier des masques RGB
img_dir = "/content/drive/MyDrive/classification_dataset/projet/dataset_COCO/train/augm_img"   # Dossier des images originales
output_dir = "/content/drive/MyDrive/classification_dataset/projet/dataset_COCO/dataset_yolo"
split = "train"  # ou "val"
```

```python
color_to_class = {
    (31, 119, 180): 0,
    (44, 160, 44): 1,
    (148, 103, 189): 2,
    (23, 190, 207): 3,
    (188, 189, 34): 4
}
```

```python
os.makedirs(f"{output_dir}/images/{split}", exist_ok=True)
os.makedirs(f"{output_dir}/labels/{split}", exist_ok=True)
```

```python
for mask_name in os.listdir(mask_dir):
    if not mask_name.endswith(('.png', '.jpg')): continue

    # Charger masque
    mask_path = os.path.join(mask_dir, mask_name)
    mask = np.array(Image.open(mask_path))[:, :, :3]  # ignore alpha

    # Trouver les couleurs présentes
    unique_colors = np.unique(mask.reshape(-1, 3), axis=0)
    classes_present = set()

    for color in unique_colors:
        color_tuple = tuple(color)
        if color_tuple in color_to_class:
            classes_present.add(color_to_class[color_tuple])

    # Écrire fichier .txt
    label_file = os.path.join(f"{output_dir}/labels/{split}", mask_name.replace(".png", ".txt").replace(".jpg", ".txt"))
    with open(label_file, "w") as f:
        f.write(" ".join(str(cls) for cls in sorted(classes_present)))

    # Copier image correspondante
    image_path = os.path.join(img_dir, mask_name)
    target_path = os.path.join(f"{output_dir}/images/{split}", mask_name)
    if os.path.exists(image_path):
        shutil.copyfile(image_path, target_path)

print("✅ Conversion terminée.")
```
